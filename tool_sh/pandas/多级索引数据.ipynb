```python
def add_info_for_mul_index_df(sv_stat, sample_info):
    name = sv_stat.split('/')[-1].split('.')[0]
    # read SV stat file
    raw_stat = pd.read_excel(sv_stat, header=[0,1], index_col=0)

    # read the sample info file
    info = pd.read_csv(sample_info, sep='\t', header=None, names=[ 'depth', 'read length'], index_col=0)

    multi_index = pd.MultiIndex.from_tuples([('Information', 'depth'), ('Information', 'read length')])

    info.columns = multi_index

    # merge the two dataframes
    df3 = pd.merge( info, raw_stat,left_index=True, right_index=True, how='left')

    # save the merged dataframe
    df3.to_excel(name + '_with_info.xlsx')
    
add_info_for_mul_index_df('./raw_sv_stat.xlsx', './all_bam.sortbydepth.info')
add_info_for_mul_index_df('./filtered_sv_stat.xlsx', './all_bam.sortbydepth.info')
```

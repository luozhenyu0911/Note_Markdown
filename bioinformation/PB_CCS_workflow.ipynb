### PacBio HiFi 质控，比对和SVs变异检测方面的相关软件
- 1. 质控软件主要有：
  - FastQC：适用于快速检查碱基质量分布、GC含量、序列长度分布等。
  - [NanoPlot](https://github.com/wdecoster/NanoPlot)：专门用于长读长数据，能够生成详细的读长分布、质量分布等可视化图表。最近更新2023.10.29
  - [Nanofilt](https://github.com/wdecoster/nanofilt): 长读取测序数据的过滤和修整。最近更新 2022.12.21
  - [chopper](https://github.com/wdecoster/chopper/)： Rust implementation of NanoFilt+NanoLyse, both originally written in Python. This tool, intended for long read sequencing such as PacBio or ONT, filters and trims a fastq file. 最近更新 2024.08.20

- 2. 比对软件主要有：
  - [minimap2](https://github.com/lh3/minimap2)：最近更新2024.03.27
  - [NGMLR](https://github.com/philres/ngmlr): 文章于2018年发表于Nature Methods，最近更新2018.06.25
  - [PBmm2](https://github.com/PacificBiosciences/pbmm2)：是PB官方基于minimap2进行优化的版本，最近更新2024.10.30
  - [Winnowmap](https://github.com/marbl/Winnowmap)：一个基于minimap2改进的映射工具，它采用了独特的优化策略，尤其是在处理复杂重复区域时，如人类染色体中的长串联重复序列。最近更新2021.05.08

- 3. SVs变异检测软件主要有：
  - [PBSV](https://github.com/PacificBiosciences/pbsv)：PacBio官方开发的结构变异软件，最近更新2024.10.29
  - [Sniffles2](https://github.com/fritzsedlazeck/Sniffles)，版本一于2018年发表于Nature methods，最近更新2024.12.17
  - [cuteSV2](https://github.com/tjiangHIT/cuteSV)，文章于2020年发表于Genome Biology，最近更新2024.05.10
  - [PAV](https://github.com/EichlerLab/pav)，Phased Assembly Variant Caller，文章于2021年发表于Science，最近更新2024.12.18
  - [DeBreak](https://github.com/Maggi-Chen/DeBreak)，文章于2023年发表在Nature Communication上，最近更新2022.12.01
  - [SVIM](https://github.com/eldariont/svim)，文章于2019年发表于Bioinformatics，最近更新2021.06.18
  - [SVision](https://github.com/xjtu-omics/SVision), 文章于2022年发表于Nature Methods，最近更新2024.05.08

### 2024 NC Tradeoffs in alignment and assembly-based methods for structural variant detection with long-read sequencing data
- https://doi.org/10.1038/s41467-024-46614-z
- 文章首先系统地分析和评估了 2014 年至 2022 年间在大量 PacBio 和 ONT 数据集上引入的 16 种最先进的长读长 SV 方法的性能。
- 不同比对工具： 
  - a） 不同的对准器对基于对齐的工具的插入召回率有显着影响。一般来说，对于大多数基于比对的工具，minimap2、Winnowmap 和 LRA 在插入召回方面优于 NGMLR。
  - b） 插入精度、删除召回率和精度不受影响，或者仅受不同对准器的细微影响
- 不同的SV caller： 
  - a） 对于易位，在模拟数据上，总体最佳工具是 cuteSV 和 pbsv（在所有合适的数据集上 F1 > 0.95），其次是 NanoSV 和 SVIM。
  - b） 对于倒置，没有工具达到高性能。NanoVar、Sniffles2 和 NanoVar 分别在 Hifi、CLR 和 ONT 上获得了最好的 F1 分数。
  - c） 对于重复，pbsv 是 Hifi 和 CLR 数据的最佳工具，其次是 DeBreak，而 NanoVar 是 ONT 数据的最佳工具。
